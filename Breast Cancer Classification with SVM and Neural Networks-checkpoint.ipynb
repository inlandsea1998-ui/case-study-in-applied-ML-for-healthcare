```python
import os
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

```python
data = load_breast_cancer(return_X_y=True, as_frame=True)
X = data[0]
y = data[1]
print(X.shape)
print(y.shape)
X_train, X_test, y_train, y_test = \
train_test_split(X, y, test_size=0.2, random_state=24)
```

```python
#SVM
from sklearn.model_selection import GridSearchCV

param_grid = [{'kernel': ['linear','sigmoid','rbf'],
               'C':[1,10,20,50,100]}]

best_grid = GridSearchCV(SVC(probability=True), param_grid, verbose=1)
best_svm_model = best_grid.fit(X_train, y_train)
print(best_grid.best_params_)
pred = best_svm_model.predict(X_test)
print(confusion_matrix(y_test,pred))
best_svm_model.score(X_test, y_test)
```

```python
pred_prob = best_svm_model.predict_proba(X_test)
preds = pred_prob[:,1]
fpr, tpr, threshold = metrics.roc_curve(y_test, preds)
roc_auc = metrics.auc(fpr, tpr)
print(pred_prob[0:10])
```

```python
import matplotlib.pyplot as plt
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
#Neural Networks
model_nn = MLPClassifier(random_state=24)
model_nn.fit(X_train, y_train)
pred = model_nn.predict(X_test)
print(confusion_matrix(y_test,pred))
print(model_nn.score(X_test,y_test))
```

```python
pred_prob = model_nn.predict_proba(X_test)
preds = pred_prob[:,1]
fpr, tpr, threshold = metrics.roc_curve(y_test, preds)
roc_auc = metrics.auc(fpr, tpr)
```

```python
plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
